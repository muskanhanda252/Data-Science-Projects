```python
def black_canvas (h,w):
    b_image = np.zeros((h, w, 3),np.uint8) #using np.zeros to create a matrix full of 0s, that represents black colour.
    return b_image
```

```python
canvas1 = black_canvas(430,430)   #creating canvas for the outer layer
canvas_center = black_canvas(120,120)#creating canvas for the center image
plt.imshow(canvas1)
plt.show()
plt.imshow(canvas_center)
plt.show()
```

```python
canvas_center[10:110,10:110] = center
plt.imshow(canvas_center)
plt.show()
```

```python
canvas1[10:210,10:210] = top_left
canvas1[10:210,220:420] = top_right
canvas1[220:420,10:210] = bt_left
canvas1[220:420,220:420] = bt_right
canvas1[155:275,155:275] = canvas_center
plt.imshow(canvas1)
plt.show()
```

```python
r_channel = canvas1[:,:,0]
g_channel = canvas1[:,:,1]
b_channel = canvas1[:,:,2]

r_channel = r_channel.reshape((430*430,1))
g_channel = g_channel.reshape((430*430,1))
b_channel = b_channel.reshape((430*430,1))
img = np.hstack((r_channel,g_channel,b_channel))
```

```python
print(img.shape)

df = pd.DataFrame(img,columns=['r','g','b'])

df.to_csv("Done.csv",index=False)
```
